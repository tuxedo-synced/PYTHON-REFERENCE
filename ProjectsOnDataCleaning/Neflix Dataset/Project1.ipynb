```python
# rating coloumn is irrelavent so we can drop it
data.drop(columns=["rating"]) # if want to drop multiple coloumns just keep a ',' and place the column .
# The above  change is not permanent so for chsning permanently we have 2 methods 
# 1. adding a parameter inplace and set it to True
# 2. equate it with data variable
```

```python
data = data.drop(columns=["rating"])
```

```python
data.drop(columns=["rating"],inplace=True)
```

```python
data.head()
```

```python
data.columns
```

```python
# To capitalize the columns :-
new_column_names = []
for every_column in data.columns:
    new_column_names.append(every_column.capitalize())
data.columns = new_column_names
data.columns
```

```python
data.head()
```

```python
# True → this row/value is a duplicate .
# False → this is the first occurrence .
data.duplicated()
```

```python
# for cleaning duplicates 
data.drop_duplicates(inplace = True)
data.head()
```

```python
data.isna()
```

```python
data.isna().sum()   #If sum is zero then there is no NaN values .
```

```python
data.dropna(inplace=True)
data.head()
```

```python
data["Show_id"].unique()
```

```python
data["Show_id"].nunique()
```

```python
data["Show_id"] = data["Show_id"].apply(lambda x: x.split("s")[1])
```

```python
data.head()
```

```python
data["Show_id"] = data["Show_id"].astype(int)
```

```python
data.head()
```

```python
data["Date_added"] = data["Date_added"].apply(lambda x: x.replace("/","-"))
data.head()
```

```python
data["Date_added"].apply(lambda x: x.replace("-","/"))
data.head()
```

```python
df = data["Country"].value_counts().reset_index()
df[df["Country"] == "Hong Kong"]
```
